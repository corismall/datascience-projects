```python
'''A substring in programming represents all characters from a string, between two specified 
indices. Given a variable string called seq, a student writes the following program that will
generate all nonempty substrings of seq:'''

seq = 'actacatagc'

for i in range(len(seq)+1) :     # line 1
        for j in range(i) :        # line 2
                print(seq[j:i])     # line 3
                
i=0 
while i<len(seq) :
    j=0 
    while(j<i) :
        print(seq[j:i])
        j+=1
    i+=1
                
'''Which of the following changes make the above program correct?

A. Program is correct as it is.

B. Change line 1 to:   for i in range(len(seq)+1) :

C. Change line 3 to:   print(seq[j:i+1])

D. Change line 2 to:   for j in range(i+1) :


1) B, C, and D


2) B, and C together


your answer => 3) C, and D together


your next answer => 4) Only B

'''
```
